```python
EE = Matrix([[1, 2], [1, -1]])
EE
```

```python
F=EE.exp()
F
```

```python
G = sp.simplify(F)
G
```

```python
G00 = G[0,0]
G00
```

```python
sp.simplify(G00)
```

```python
num = sp.simplify( G00 * (5*sp.sqrt(3) + 9) * sp.exp(sp.sqrt(3)) * (9- 5*sp.sqrt(3)) )
num
```

```python
denom = sp.simplify((5*sp.sqrt(3) + 9) * (9 - 5*sp.sqrt(3)) *sp.exp(sp.sqrt(3)) )
denom
```

```python
q = num / denom
q
```

```python
(q - G00).evalf()
```

```python
(q - G00).simplify()
```

```python
H = Matrix([[1, 4], [1, 1]])
H
```

```python
#The second number in each tuple is the algebraic multiplicity
eig = H.eigenvects()
pprint(eig)
```

```python
l1 = list(eig[0]); l2 = list(eig[1])
lambda_1 = l1[0]; m_1 = l1[1]; v1 = l1[2]
lambda_2 = l2[0]; m_2 = l2[1]; v2 = l2[2]
pprint(([lambda_1, m_1, v1], [lambda_2, m_2, v2]))
```
